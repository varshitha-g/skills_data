```python
import pandas as pd
```

```python
df = pd.read_excel('skill_data.xlsx')
```

```python
df.head()
```

```python
web_devloper_skills = [
    'django',
    'html',
    'css',
    'nodejs',
    'reactjs',
    'nodejs',
    'flask',
    'laravel'
]
web_devloper_skills = list(set(web_devloper_skills))

machine_learning_skills = [
    'nosql',
    'pandas',
    'numpy',
    'matpotlib',
    'seabron',
    'scipy'
]
machine_learning_skills = list(set(machine_learning_skills))
```

```python
for i, l in enumerate(df["Skills"].head()):
    print("list",i,"is",type(l))
```

```python
df["Skills"] = df["Skills"].apply(eval)
```

```python
for i, l in enumerate(df["Skills"].head()):
    print("list",i,"is",type(l), l)
```

```python
df["Skills"] = df["Skills"].apply(set)
df["Skills"] = df["Skills"].apply(list)
for i, l in enumerate(df["Skills"].head()):
    print("list",i,"is", l)
```

```python
designation  = []
no_of_skills = []
for i, l in enumerate(df["Skills"]):
    count_wd = 0
    count_ml = 0
    no_of_skills.append(len(l))
    for i in l:
        if i in machine_learning_skills:
            count_ml = count_ml + 1
        if i in web_devloper_skills:
            count_wd = count_wd + 1
    if count_ml > 2:
        designation.append('Machine Learning Engineer')
    elif count_wd > 2 :
        designation.append('Web Developer')
    else:
        designation.append('None')
print("Designation")
print(designation)
print()
print("no_of_skills")
print(no_of_skills)
```

```python
df['No_of_Skills'] = no_of_skills
df['Designation'] = designation
```
